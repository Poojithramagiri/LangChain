```python
# Import libraries and set up API key
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import ChatPromptTemplate, HumanMessagePromptTemplate, MessagesPlaceholder
from langchain.memory import ConversationBufferMemory, FileChatMessageHistory
from langchain.schema import SystemMessage

# Set up Google Gemini API credentials
api_key = "AIzaSyBCH44djBwy-dcdIT9HsI8AWrVobma-aiA"
os.environ["GOOGLE_API_KEY"] = api_key

# Initialize the LLM
llm = ChatGoogleGenerativeAI(
    api_key=api_key,
    model="gemini-1.0-pro",
    temperature=0.9
)

# Create memory
memory = ConversationBufferMemory(
    memory_key='chat_history',
    return_messages=True
)

# Set up the prompt template
prompt = ChatPromptTemplate(
    input_variables=["content", "chat_history"],
    messages=[
        SystemMessage(content="You are a chatbot having a conversation with a human."),
        MessagesPlaceholder(variable_name="chat_history"),  # Where the memory will be stored.
        HumanMessagePromptTemplate.from_template("{content}")
    ]
)

# Function to generate response
def generate_response(content, chat_history):
    # Format the prompt with the current content and chat history
    formatted_prompt = prompt.format(content=content, chat_history=chat_history)
    # Get LLM response
    response_text = llm.invoke(formatted_prompt).content
    return response_text

# Run the chat application
while True:
    content = input('Your prompt: ')
    if content.lower() in ['quit', 'exit', 'bye']:
        print('Goodbye!')
        break

    # Generate response
    response = generate_response(content, memory.chat_memory.messages)
    print(response)
    print('-' * 50)

    # Update the chat history
    memory.chat_memory.add_message({
        "role": "user",
        "content": content
    })
    memory.chat_memory.add_message({
        "role": "assistant",
        "content": response
    })
```
